```python
# Import Standard Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the Dataset
data = pd.read_csv('/content/IMDb Movies India.csv',encoding='ISO-8859-1')

# Show the first 5 lists
data.head()
```

```python
# Count rows and column
data.shape
```

```python
# Acess arrays
data.info()
```

```python
# Data cleaning

# Drop the unnecessory column
data = data.drop(['Actor 2','Actor 3'],axis=1)
data.head()
```

```python
# Count of missing values in the corresponding column of the dataframe
data.isnull().sum()
```

```python
data.isnull().sum()/len(data)*100
```

```python
# Remove duplicates values
data.drop_duplicates()
```

```python
data['Year']= data['Year'].str.replace(r'[()]','',regex=True)
data.head()
```

```python
data['Duration']= data['Duration'].str.replace(r'min','',regex=True)
data.head()
```

```python
data['Votes'].unique()
```

```python
data['Votes']= data['Votes'].str.replace(r',','',regex=True)
data.head()
```

```python
data['Genre'].value_counts()
```

```python
# Fill null volue with Drama in Genre column

data['Genre']= data['Genre'].fillna('Drama')
data.isnull().sum()
```

```python
data['Year']= data['Year'].fillna(data['Year'].mode()[0])
data.isnull().sum()
```

```python
data['Duration']= data['Duration'].fillna(data['Duration'].mode()[0])
data['Rating']= data['Rating'].fillna(data['Rating'].mean())
data['Votes']= data['Votes'].fillna(data['Votes'].mode()[0])
data['Director']= data['Director'].fillna(data['Director'].mode()[0])
data['Actor 1']= data['Actor 1'].fillna(data['Actor 1'].mode()[0])
data.isnull().sum()
```

```python
data.dtypes
```

```python
data['Duration']= data['Duration'].astype(float)
data['Year']= data['Year'].astype(int)
data['Votes']= data['Votes'].str.replace(r'M','',regex=True)
data['Votes']= data['Votes'].str.replace(r'$','',regex=True)
data['Votes']= data['Votes'].astype(str).astype(float)
data.dtypes
```

```python
# Applying the Feature Scaling and Feature Engineering
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
mms
```

```python
data[['Duration','Rating','Votes']]= mms.fit_transform(data[['Duration','Rating','Votes']])
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le
```

```python

data['Director'] = le.fit_transform(data['Director'])
data['Actor 1'] = le.fit_transform(data['Actor 1'])
data['Genre'] = le.fit_transform(data['Genre'])
data.head()
```

```python
# Applying the Machine Learing algorithm on the dataframe
# Here Rating is target variable so we apply regression algorithm

x= data.drop('Rating',axis=1)
y= data['Rating']
```

```python
# Split the dataframe into train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# Performing Prediction
y_pred = lr.predict(x_test)
y_pred
```

```python
# Evalutions Display
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
accuracy
```

```python
0.04339333837900183*100
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(lr,x,y,cv=5)
```
